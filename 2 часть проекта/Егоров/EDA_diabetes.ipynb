## Посмотрим на распределение данных относительно столбца есть диабет или нет

## Одномерный анализ признаков

````python
binary_cols = [
    "HighBP","HighChol","CholCheck","Smoker","Stroke",
    "HeartDiseaseorAttack","PhysActivity","Fruits","Veggies",
    "HvyAlcoholConsump","AnyHealthcare","NoDocbcCost",
    "DiffWalk","Sex"
]

ordinal_cols = [
    "BMI","GenHlth","MentHlth","PhysHlth","Age","Education","Income"
]
````

````python
bin_summary = df[binary_cols].mean().sort_values(ascending=False)
bin_summary
````

````python
plt.figure(figsize=(6,6))
sns.barplot(x=bin_summary.values, y=bin_summary.index)
plt.xlabel("Доля значения 1")
plt.title("Частота бинарных признаков")
plt.tight_layout()
plt.show()
````

Получили график с частотой единиц по различным бинарным признакам

````python
df.loc[:, 'BMI_log'] = np.log(df["BMI"])
df
````

````python
cols = ["BMI", "BMI_log", "MentHlth", "PhysHlth"]
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()
for ax, col in zip(axes, cols):
    sns.histplot(data=df, x=col, kde=True, ax=ax)
    ax.set_title(f"Распределение {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
````

````python
cols = ["GenHlth", "Age", "Education", "Income"]
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()
for ax, col in zip(axes, cols):
    sns.countplot(x=col, data=df, ax=ax)
    ax.set_title(f"Распределение {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
````

Посмотрели на распределения чисто вещественных признаков и порядковых признаков

## Связь признаков с таргетом

````python
group_bin = df.groupby("Diabetes_binary")[binary_cols].mean().T
group_bin.columns = ["No_diabetes", "diabetes"]
group_bin["diff"] = group_bin["diabetes"] - group_bin["No_diabetes"]
group_bin= group_bin.sort_values("diff", ascending=False)
````

````python
plt.figure(figsize=(6,6))
sns.barplot(x=group_bin["diff"],y=group_bin.index)
plt.axvline(0, color="grey", linestyle="--")
plt.xlabel("Разница доли (диабет - без диабета)")
plt.ylabel("Сами разницы долей")
plt.title("Как бинарные признаки отличаются у пациентов с диабетом")
plt.tight_layout()
plt.show()
````

Построили разницу между долями встречь единиц у двух групп людей

````python
for col in ["BMI","MentHlth","PhysHlth"]:
    plt.figure(figsize=(5,4))
    sns.boxplot(x="Diabetes_binary", y=col, data=df)
    plt.title(f"{col} в группах с/без диабета")
    plt.tight_layout()
    plt.show()
````

Boxplot-ы по вещественным признакам при делений групп на людей с диабетом и нет

## Корреляции между признаками

````python
num_for_corr = ["Diabetes_binary","BMI","GenHlth","MentHlth",
                "PhysHlth","Age","Education","Income"]

corr = df[num_for_corr].corr(method="spearman")

plt.figure(figsize=(7,6))
sns.heatmap(corr, annot=True, fmt=".2f", square=True)
plt.title("Spearman корреляции (включая Diabetes_binary)")
plt.tight_layout()
plt.show()
````

# Feature engineering

## 1) Флаг по ожирению

````python
df.loc[:, "Obese_flag"] = (df["BMI"] >= 30).astype(int)

df["Obese_flag"].value_counts()
````

## 2) Категории по Возростам

````python
age_group_map = {
    1: 1, 2: 1, 3: 1,        # 18–34
    4: 2, 5: 2, 6: 2, 7: 2,  # 35–54
    8: 3, 9: 3,              # 55–64
    10: 4, 11: 4, 12: 4, 13: 4  # 65+
}

df.loc[:, "Age_group_num"] = df["Age"].map(age_group_map)
df
````


```
# 18–34 молодые люди
```

```
# 35–54 средний возраст
```
```
# 55–64 средний+ возраст
```
```
# 65+ пожилые люди
```


## 3) Флаг по доходу

````python
df.loc[:, "LowIncome_flag"] = (df["Income"]<=3).astype(int)

df["LowIncome_flag"].value_counts()
````

````python
df.loc[:, "HighIncome_flag"] = (df["Income"]>=7).astype(int)

df["HighIncome_flag"].value_counts()
````

## 4) Индекс по социальному уровню

````python
df['soc_index']= df['Income']+ df['Education']
````

## 5) Индекс по образу жизни

````python
df.loc[:, "NonSmoker_flag"] = (df["Smoker"]==0).astype(int)
````

````python
df.loc[:, "NoHeavyAlcohol_flag"] = (df["HvyAlcoholConsump"]==0).astype(int)
````

````python
df.loc[:, "HealthyDiet_flag"] = ((df["Fruits"]==1) & (df["Veggies"]==1)).astype(int)
````

````python
df['Lifestyle_score']= df['NoHeavyAlcohol_flag']+df['NonSmoker_flag']+df['HealthyDiet_flag']+ df['PhysActivity']
df["Lifestyle_score"].value_counts()
````

## 6) Индекс сердечным проблемам

````python
df.loc[:, "NoHighBP"] = (df["HighBP"]==0).astype(int)
df.loc[:, "NoHighChol"] = (df["HighChol"]==0).astype(int)
df.loc[:, "NoStroke"] = (df["Stroke"]==0).astype(int)
df.loc[:, "NoHeartDiseaseorAttack"] = (df["HeartDiseaseorAttack"]==0).astype(int)
df.loc[:, "NoDiffWalk"] = (df["DiffWalk"]==0).astype(int)
df
````

````python
df['cardio_index']= df['NoHighBP']+ df['NoHighChol']+ df['NoStroke']+ df['NoHeartDiseaseorAttack']+ df['NoDiffWalk']
````

## 6) Индекс по человеку в целом то есть general_index

````python
df['General_index']= df['cardio_index']+ df['Lifestyle_score']+ df['soc_index']
df['General_index'].value_counts()
````

````python
df = df.drop(columns=["NoHeartDiseaseorAttack", "NoStroke", "NoDiffWalk", 'NoHighChol', 'NoHighBP', 'NonSmoker_flag', 'NoHeavyAlcohol_flag', 'HealthyDiet_flag'])
````
